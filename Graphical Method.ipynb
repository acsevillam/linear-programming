```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Construct lines
# x > 0
x = np.linspace(0, 500, 5000)
# 1/3 x1 + 1/2 x2 <= 100
y1 = 2*(100 - 1/3*x)
# 1/3 x1 + 1/6 x2 <= 80
y2 = 6*(80 - 1/3*x)
# x2 <= 150
y3 = 150*np.ones(5000)
# x2 >=0 
zeros=np.zeros(5000)
```

```python
# Make plot

fig=plt.figure(figsize=(8,8))
ax1=fig.add_subplot(111)
ax1.tick_params(labelsize=12)

ax1.plot(x, y1, label=r'$ 1/3 x_{1} + 1/2 x_{2} \leq 100$')
ax1.plot(x, y2, label=r'$ 1/3 x_{1} + 1/6 x_{2} \leq 80$')
ax1.plot(x, y3, label=r'$ x_{2} \leq 150$')
ax1.axvline(x=0, label=r'$ x_{1} \geq 0$', color='purple', linewidth=5)
ax1.plot(x, zeros, label=r'$ x_{2} \geq 0$', color='red', linewidth=5)
ax1.set_xlim((0, 300))
ax1.set_ylim((0, 200))
ax1.set_xlabel(r'$x_{1}$',fontsize=18,labelpad=18)
ax1.set_ylabel(r'$x_{2}$',fontsize=18,labelpad=18)

# Fill feasible region
feasible_max = np.minimum(y1, y2)
feasible_max = np.minimum(feasible_max, y3)
ax1.fill_between(x,feasible_max, color='grey', alpha=0.5)
ax1.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#Se define la granuladidad de las celdas (que tan pequenias son las celdas)
#Maximo y minimo del eje x
xmax = 300
xmin = 0
#Cuantos puntos se buscan, por eje x o y
NX = 15
NY = 15
#Maximo y minimo del eje y
ymax = 300
ymin = 0
#Se crea la red y se calculan los componentes del vector
x_mesh = np.linspace(xmin, xmax, NX)
y_mesh = np.linspace(ymin, ymax, NY)
 
#Con meshgrid creamos 2 redes donde 'x' y 'y' contiennen las coordenadas de 'x' y 'y' por cada punto.
X, Y = np.meshgrid(x_mesh, y_mesh)
#Estas son las ecuaciones diferenciales, que son sustituidos por sus respectivos valores donde
# es para x y By es para y
Bx = 15
By = 10
```

```python
# Make plot
fig=plt.figure(figsize=(8,8))
ax1=fig.add_subplot(111)
ax1.tick_params(labelsize=12)

ax1.plot(x, y1, label=r'$ 1/3 x_{1} + 1/2 x_{2} \leq 100$')
ax1.plot(x, y2, label=r'$ 1/3 x_{1} + 1/6 x_{2} \leq 80$')
ax1.plot(x, y3, label=r'$ x_{2} \leq 150$')
ax1.axvline(x=0, label=r'$ x_{1} \geq 0$', color='purple',linewidth=5)
ax1.plot(x, zeros, label=r'$ x_{2} \geq 0$', color='red',linewidth=5)
ax1.quiver(X,Y,Bx,By,label=r'$\nabla f(x_{1},x_{2})$', color='grey', alpha=0.5)
ax1.set_xlim((0, 300))
ax1.set_ylim((0, 200))
ax1.set_xlabel(r'$x_{1}$',fontsize=18,labelpad=18)
ax1.set_ylabel(r'$x_{2}$',fontsize=18,labelpad=18)

# Fill feasible region
feasible_max = np.minimum(y1, y2)
feasible_max = np.minimum(feasible_max, y3)
ax1.fill_between(x,feasible_max, color='grey', alpha=0.5)
ax1.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
